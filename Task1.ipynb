# Get Our Workspace Ready

* Import all tools needed

```python
# Data handling
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Machine learning models and tools
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

# 2. Data Exploration & Cleaning

```python
df = pd.read_csv('/content/drive/MyDrive/ML intern/Task1/StudentPerformanceFactors.csv')
```

```python
df.head()
```

```python
# Understandin the dataset
print(df.info())
print(df.describe())
print(df.isnull().sum())
print(df.shape)
```

```python
df_clean = df.drop(columns=['Teacher_Quality', 'Parental_Education_Level', 'Distance_from_Home'])
df_clean.shape
```

```python
x = df_clean[["Hours_Studied"]]
y = df_clean["Exam_Score"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

### 3. Building the Model

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_predict = model.predict(x_test)
```

### 4. Evaluation

```python
print("mean squared error= ", mean_squared_error(y_test,y_predict))
print("r2 score= ", r2_score(y_test,y_predict))
```

###5. Data Visualization

```python
plt.figure(figsize=(8,5))
plt.scatter(x, y , color='red', edgecolor='black')
plt.xlabel("Hours studied")
plt.ylabel("Exam Score")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(y_test, y_predict, color="green", edgecolor="black")
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r')
plt.xlabel("Actual exam score")
plt.ylabel("Predicted Exam score")
plt.title("Actual vs Predicted")
```

# Bonus Section:

Try Polynoimal Regression and compare performance

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2)
x_poly_train =  poly.fit_transform(x_train)
x_poly_test = poly.transform(x_test)
```

```python
poly_model = LinearRegression()
poly_model.fit(x_poly_train, y_train)
y_pred_poly = poly_model.predict(x_poly_test)
```

```python
# Performance Metrics
print(" Linear Regression:")
print("MSE: ", mean_squared_error(y_test, y_predict))
print("R2 Score: ", r2_score(y_test, y_predict))

print("Polynoimal Regression: ")
print("MSE: ", mean_squared_error(y_test, y_pred_poly))
print("R2 Score: ", r2_score(y_test, y_pred_poly))
```

# Comparing Visualization between Linear and Polynoimal Regression

```python
plt.figure(figsize=(10, 5))

# Scatter the original data points
plt.scatter(x, y, color='grey', edgecolors='black', label='Actual Data')

# Create a smooth range of input values for plotting
x_range = np.linspace(x.min(), x.max(), 100).reshape(-1, 1)
x_range_df = pd.DataFrame(x_range, columns=["Hours_Studied"])

# Make predictions
y_linear_plot = model.predict(x_range_df)  # OK with feature names
y_poly_plot = poly_model.predict(poly.transform(x_range_df.values))  # FIX: remove feature names

# Plot linear regression line
plt.plot(x_range, y_linear_plot, color='blue', label='Linear Regression')

# Plot polynomial regression curve
plt.plot(x_range, y_poly_plot, color='red', label='Polynomial Regression')

# Labeling
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.title("Linear vs Polynomial Regression")
plt.legend()
plt.grid(True)
plt.show()
```
